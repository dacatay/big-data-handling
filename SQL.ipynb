# SQL

SQL ist eine Datenbanksprache zur Definition von Datenstrukturen in relationalen Datenbanken sowie zum Bearbeiten (Einfügen, Verändern, Löschen) und Abfragen von darauf basierenden Datenbeständen.<br>
<br>
Man kann SQL-Datenbanken in Azure oder mit AWS hosten und so vielen Leuten Zugriff auf die Daten gewähren.<br>
Python und R bieten viele Pakete an um SQL Abfragen zu generieren(Dask und Pandas).<br>
SQL Datenbanken sind immer ähnlich wie Dataframes aufgebaut in Zeilen und Spalten.<br>
SQL ist eine Sprache um die Datenbanken abzufragen.<br>


## SQL Syntax 

select (gebe Spalten der Tabelle an die du willst oder * für alle Spalte)<br>
from (wähle die Tabelle in der Datenbank aus)<br>
where (Eine Einschränkung der Daten wie z.B. <100)<br>
inner join (wähle eine Tabelle mit der gejoint werden soll) on (wähle Spalten für den join in beiden Tabellen)<br>
group by (erstelle Gruppen von einer Spalte)<br>

```python
from sqlalchemy import create_engine # database connection
import pandas as pd
import time
```

Erstellen einer Datenbank mit SQLite. SQLite ist für lokale Datenbanken gebaut, im Gegensatz zu SQL Servern die irgendwo gehostet werden.

```python
disk_engine = create_engine('sqlite:///taxi.db')
```

chunksize = Datenmenge auf einmal eingeladen in 10.000 Spalten

```python
start = time.time()
chunksize = 10000
j = 0
index_start = 1
```

```python
# loop durch die Datenmenge/CSV Zeile
# man kann auch eine zweite Datenbank mit nur 1000 Zeilen anlegen um die Daten in dieser DB anzugucken
```

```python
for df in pd.read_csv(r'taxi_hunderttausend.csv', delimiter = ',', chunksize=chunksize,iterator=True,error_bad_lines=False):
    #index anknüpfen
    df.index += index_start
    j+=1
    #Information über Menge und Zeit
    print('{} seconds:', start-time.time(),' completed {} rows', j*chunksize)
    df.to_sql('taxi', disk_engine, if_exists='append')
    index_start = df.index[-1] + 1    
```

SQL Datenbanken können als Dataframes in Dask und Pandas eingeladen werden.
Pandas kann auch SQL Befehle ausführen

```python
expr = 'select * from taxi'
df = pd.read_sql_query(expr, disk_engine)
df
```

```python
expr = '''select  pickup_datetime,dropoff_datetime,Pickup_longitude,Pickup_latitude,Dropoff_longitude,Dropoff_latitude
from taxi
where Tip_amount >5'''
df = pd.read_sql_query(expr, disk_engine)
df
```

# Blaze

The Blaze Ecosystem provides Python users high-level access to efficient computation on inconveniently large data. Blaze can refer to both a particular library as well as an ecosystem of related projects that have spun off of Blaze development.

Blaze kann verschiedene Datentypen(CSV,HDF,SQL,...) mit der selben Abfrage bearbeiten

```python
from blaze import data, by, compute
```

Data ist eine Bibliothek um Datenquellen zu öffnen. SQL,CSV...

```python
table = data("sqlite:///taxi.db::taxi")
```

```python
print(table.peek())
```

Beispiel suche den sql Befehl für eine Pandas Abfrage

```python
expr = table[table.Payment_type<20]
print(compute(expr))
```

```python
expr = by(table.Payment_type, amount=table.Total_amount.mean())
print(compute(expr))
```

Blaze ist ein Interface um verschiedene Datenbanken/Speicher abzufragen. Es geht weniger darum wie die Daten im Speicher verarbeitet werden, sondern wie kann man alle möglichen Datentypen mit einem Pakage benutzen.
Blaze beinhaltet auch Dask für out of Ram Computation.
